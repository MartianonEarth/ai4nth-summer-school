```python
import importlib.metadata
from IPython import get_ipython

# Your list of dependencies
dependencies = [
    "equinox",
    "ipympl",
    "jax",
    "jupyter",
    "matplotlib",
    "optax",
    "pandas",
    "rich",
    "seaborn"
]

missing_packages = []

print(f"{'Library':<15} | {'Status':<15}")
print("-" * 35)

# 1. Check for installed packages
for lib in dependencies:
    try:
        version = importlib.metadata.version(lib)
        print(f"{lib:<15} | Installed ({version})")
    except importlib.metadata.PackageNotFoundError:
        print(f"{lib:<15} | NOT Installed")
        missing_packages.append(lib)

# 2. Install missing packages if any are found
if missing_packages:
    print(f"\nFound {len(missing_packages)} missing package(s). Installing now...")
    
    # Get the active Jupyter environment
    ipython = get_ipython()
    
    # Construct the conda arguments (e.g., install -c conda-forge -y pkg1 pkg2)
    conda_args = f"install -c conda-forge -y {' '.join(missing_packages)}"
    
    # Execute the %conda magic command dynamically
    ipython.run_line_magic('conda', conda_args)
    
    print("\n✅ Installation complete! NOTE: You may need to restart your Jupyter kernel to use them.")
else:
    print("\n✅ All packages are already installed!")
```
